The purpose of this notebook is to pilot the code for reformatting the dataset for gpp

```r
#Set up the environment
library(tidyverse)
library(reshape2)
```

```r
#Load in the example dataset and format it
setwd("/home/jupyter/Ancestry/analysis/reformat_gpp_dataset")
genotype.dataset = read.table("chrY.INFO.txt", sep = "\t") %>%
select(1:15) %>%
rename(
    "chr" = 1,
    "pos" = 2,
    "ref" = 3,
    "alt" = 4,
    "af_total" = 5,
    "af_oth" = 6,
    "af_ami" = 7,
    "af_sas" = 8,
    "af_fin" = 9,
    "af_eas" = 10,
    "af_asj" = 11,
    "af_amr" = 12,
    "af_afr" = 13,
    "af_mid" = 14,
    "af_nfe" = 15
) %>%
mutate_at(vars(5:15), as.numeric) %>%
rowwise() %>%
mutate("max_subpopulation_af" = max(c_across(6:15), na.rm = TRUE)) %>%
filter(!max_subpopulation_af %in% -Inf) %>% #Remove the infinite values
mutate("keep_drop" = ifelse(af_total > 0.01 | max_subpopulation_af > 0.02, "keep", "drop")) %>%
filter(keep_drop %in% "keep") %>%
select(-keep_drop)

#Write it out
write.table(genotype.dataset, "", sep = "\t", col.names = TRUE, row.names = FALSE, quote = FALSE)

genotype.dataset %>% head()
```
